```python
from Bio import SeqIO
from Bio.Seq import Seq
human_dir={}
records = SeqIO.parse("./hg19.fa", "fasta")
for record in records:
    id = record.id
    seq = record.seq
    human_dir[id] = seq
```

```python
file = open('H:/002.singcell/01.liver/10.mutileSpecies/05.SAMap/015.enhancer/hg19.fa.fai','r')
chr_info = {}
for line in file:
    line = line.strip().split(',')[0].split(' ')
    nc = line[0][1: ]
    chr =line[-1]
    if chr != 'mitochondrion':
        id = 'chr' + chr
        chr_info[id] = nc
```

```python
tmp =[]
table = open('./liver_enhancer.fa','w')
for index,row in df_liver.iterrows():
    name = row['name']
    start = row['start']
    end = row['end']
    chr = row['chr']
    id = name + '|' + 'chr'+ "_" + str(start) + '_' + str(end)
    seq = human_dir[chr_info[chr]][start-1:end]
    if id not in tmp:
        table.write('>'+id+'\n')
        table.write(str(seq).upper()+'\n')
        if len(str(seq)) == 0:
            print(name)
        tmp.append(id)
    else:
        pass
table.close()
```

```python
import os,re,sys
from Bio import SeqIO
from Bio.Seq import Seq
import pandas as pd
human_dir={}
records = SeqIO.parse("./mouse.fa", "fasta")
for record in records:
    id = record.id
    seq = record.seq
    human_dir[id] = seq

df = pd.read_csv('scACAT.txt',sep='\t',header=None,names=['id','deepth'],index_col=None)
set([x.split(":")[0] for x in df['id']])
result = open('scACAT.fa','w')
for i in df['id']:
    if i=='MT':
        continue
    else:
        chr = re.search(r'chr(\w+)',i.split(':')[0]).group(1)
        start = int(i.split(':')[1].split('-')[0])
        end = int(i.split(':')[1].split('-')[1])
        seq = human_dir[chr][start:end]
        result.write('>'+i+'\n'+str(seq)+'\n')
result.close()
```

```python
import os,re,sys
import pandas as pd
def gene2names(gene_file,name_flie):
    gene = pd.read_csv(gene_file,sep='\t',header=0,index_col=None)
    name = pd.read_csv(name_flie,sep='\t',header=None,index_col=None,names=['gene_id','name','expression'])
    name_dic = name.set_index('gene_id').to_dict()['name']
    gene['gene_name'] = gene['gene_id'].map(name_dic)
    gene.to_csv(f'{sys.argv[0]}.gene_name',sep='\t',header=True,index=False)
    return None
gene2names('08.peak_annotation.py.out','lamprey.tsv')
```
